We can take care of the missing values in our dataset 

```python
df.isna()
```

```python
df.isna().sum()
```

```python
# to check the percentage of missing data
missing_per = df.isna().sum()*100/len(df)
print(missing_per)
```

```python
df['__year_resale_value'].fillna(method = 'ffill', inplace = True)
Price_mean=df['Price_in_thousands'].mean()
df['Price_in_thousands'].fillna(Price_mean, inplace = True)
Engine_mean = df['Engine_size'].mean()
df['Engine_size'].fillna(Engine_mean, inplace= True)
horse_mean=df['Horsepower'].mean()
df['Horsepower'].fillna(horse_mean, inplace = True)
df['Fuel_efficiency'].fillna(method = 'bfill', inplace= True)
df['Wheelbase'].fillna(method = 'ffill', inplace=True)
df['Width'].fillna(method='ffill', inplace=True)
df['Length'].fillna(method='bfill', inplace=True)
curb_mean = df['Curb_weight'].mean()
df['Curb_weight'].fillna(curb_mean, inplace=True)
fuel_mean = df['Fuel_capacity'].mean()
df['Fuel_capacity'].fillna(fuel_mean, inplace=True)
df['Fuel_efficiency'].fillna(method='bfill', inplace=True)
df['Power_perf_factor'].fillna(method='ffill', inplace=True)
```

I have used various ways to illustrate how one can eliminate missing values in the dataset 

```python
# we confirm that all the missing values are dealt with
df.isna().sum()
```
